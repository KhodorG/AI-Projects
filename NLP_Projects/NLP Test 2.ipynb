```python
def Mic():
    re = sr.Recognizer()
    with sr.Microphone() as source:
        print("Brainly is listening...")
        wav = re.listen(source)
        try:
            inp = re.recognize_google(wav)
            print("You:")
            print(inp)
            inp = inp.upper()
            print("Brainly:")
            if (inp in Greetings):
                rd=random.randint(0,len(brainy_response)-1)
                print(brainy_response[rd])
            elif (inp in Questions[0]):
                print(lines[0])
            elif (inp in Questions[1] or inp in Questions[2]):
                print(lines[2])
            elif (inp in Questions[3]):
                print(lines[4])
            else :
                print("Sorry didn't know what you are saying")
        except:
            print("Error")
       
    
        
```

```python
while (True):
    mic_input = Mic()
    if (mic_input in leave):
        break
        
```

```python
def recorded(audio_path):
    r = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio = r.record(source)
    inp = r.recognize_google(audio)
    print("You:")
    print(inp)
    inp = inp.upper()
    print("Brainly:")
    if (inp in Greetings):
        rd=random.randint(0,len(brainy_response)-1)
        print(brainy_response[rd])
    elif (inp in leave):
        print("")
    elif (inp in Questions[0]):
        print(lines[0])
    elif (inp in Questions[1] or inp in Questions[2]):
        print(lines[2])
    elif (inp in Questions[3]):
        print(lines[4])
    
```
